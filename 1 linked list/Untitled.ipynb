```python
class Node:
    def __init__(self, data,nextNode=None):
        self.data = data
        self.next = None
        self.nextNode = nextNode
    def getNextNode(self):
        return self.nextNode
    def getData(self):
        return self.data
    
def findNode(self,value):
    curr = self.head
    while curr: 
        if curr.getData() == value:
               return True
        curr = curr.getNextNode()
    return False

def delete(head, value):
    
    prev = None
    curr = head
    while curr:
        if curr.getData() == value:
            if prev:
                prev.setNextNode(curr.getNextNode())
            else:
                head = curr.getNextNode()
            return True
        prev = curr
        curr = curr.getNextNode() 
    return head
    return False
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()

# Main
# Read the link list elements including -1
arr=list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
l = ll(arr[:-1])
value=int(input())
l = delete(l, value)
printll(l)
```
